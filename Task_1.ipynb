```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
project_path = os.getcwd()
file_path = project_path +'/task_1.csv'
df = pd.read_csv(file_path,encoding='ISO-8859-1')
df
```

# Data Understanding, cleaning , Formating

```python
print("Dataset Shape is : ",df.shape)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df
```

```python
df.dtypes
```

```python
df.dropna(subset=['CustomerID'])
```

```python
df = df[df['Quantity']>0]
```

```python
df = df.copy()
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df = df[df['UnitPrice'] > 0]
df = df.copy()
```

```python
print("data shape is :",df.shape)
df.describe()
```

# Data Analysis

```python
monthly_revenue = df.groupby('YearMonth')['Revenue'].sum().reset_index()
monthly_revenue.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_revenue['YearMonth'].astype(str),
         monthly_revenue['Revenue'])

plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Year-Month")
plt.ylabel("Total Revenue")
plt.show()
```

# Top-Selling Products Analysis

```python
top_products = df.groupby('Description')['Revenue'].sum().reset_index()

top_products = top_products.sort_values(by='Revenue', ascending=False)

# Top 10 products
top_10_products = top_products.head(10)

top_10_products
```

```python
plt.figure(figsize=(12,6))
plt.barh(top_10_products['Description'],
         top_10_products['Revenue'])

plt.xlabel("Total Revenue")
plt.title("Top 10 Revenue Generating Products")
plt.gca().invert_yaxis()
plt.show()
```

# Country Performance Analysis

```python
country_revenue = df.groupby('Country')['Revenue'].sum().reset_index()

country_revenue = country_revenue.sort_values(by='Revenue', ascending=False)

top_countries = country_revenue.head(10)
```

```python
plt.figure(figsize=(12,6))
plt.barh(top_countries['Country'],
         top_countries['Revenue'])

plt.xlabel("Total Revenue")
plt.title("Top 10 Revenue Generating Countries")
plt.gca().invert_yaxis()
plt.show()
```

# Customer Analysis

```python
total_customers = df['CustomerID'].nunique()
print("Total Unique Customers:", total_customers)
```

```python
customer_revenue = df.groupby('CustomerID')['Revenue'].sum().reset_index()

customer_revenue = customer_revenue.sort_values(by='Revenue', ascending=False)

top_10_customers = customer_revenue.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(top_10_customers['CustomerID'].astype(str),
        top_10_customers['Revenue'])

plt.xlabel("Customer ID")
plt.ylabel("Total Revenue")
plt.title("Top 10 Revenue Generating Customers")
plt.show()
```

# Customer Behavior Analysis

```python
total_orders = df['InvoiceNo'].nunique()
print("Total Orders:", total_orders)
```

```python
# Revenue per Order
order_revenue = df.groupby('InvoiceNo')['Revenue'].sum()

average_order_value = order_revenue.mean()

print("Average Order Value:", round(average_order_value, 2))
```

```python
customer_orders = df.groupby('CustomerID')['InvoiceNo'].nunique()

customer_orders.describe()
```

```python
filtered_orders = customer_orders[customer_orders < 50]

plt.figure(figsize=(8,5))
plt.hist(filtered_orders, bins=30)
plt.title("Customer Purchase Frequency (Without Extreme Outliers)")
plt.xlabel("Number of Orders")
plt.ylabel("Number of Customers")
plt.show()
```
